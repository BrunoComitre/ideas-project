```python
class MyListener(AbstractEventListener):
    def before_navigate_to(self, url, driver):
        print("Antes de abrir a url %s" % url)

    def after_navigate_to(self, url, driver):
        print("Depois de abrir a url %s" % url)

    def before_click(self, element, driver):
        print("Antes de clicar no elemento")

    def after_click(self, element, driver):
        print("Depois de clicar no elemento")

    def before_close(self, driver):
        print("Antes de fechar a pagina")

    def after_close(self, driver):
        print("Depois de fechar a pagina")

    def on_exception(self, exeception, driver):
        print("Ocorreu um erro")


if __name__ == '__main__':
    
    firefox = webdriver.Firefox()
    
    firefox.maximize_window() 

    listener = MyListener()

    ef_firefox = EventFiringWebDriver(firefox, listener)

    ef_firefox.get('http://google.com.br/')
    
    campo_busca = ef_firefox.find_element_by_name('q')

    campo_busca.send_keys('Kaggle')

    campo_busca.send_keys(Keys.ENTER)
    
    campo_busca.send_keys('Kaggle: Your Home for Data Science').click()
    
    #campo_busca.find_element(By.xpath, "//div[@id='rso']/div/div/div/div/div/div/a/h3").click()
    
    campo_busca.send_keys(Keys.ENTER)
    
    ef_firefox.close()
    # Fechar navegador
    # firefox.quit()

    
# click on css=.g > div > div > .rc .LC20lb OK
# mouseOver on css=.brand-main img OK
# mouseOut on css=.brand-main img 
```
